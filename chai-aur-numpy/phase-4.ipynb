### Convert a png to np array

```python
# 1st method using pillow
from PIL import Image
img = Image.open('numpy-logo-png.png')
arr = np.array(img)
print(arr.shape)
print(arr.dtype)
np.save('numpy-logo.npy',arr)
```

```python

# loading the np array created by pillow
try:
    logo1 = np.load('numpy-logo.npy')

    #Display
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo1)
    plt.title("Numpy logo")
    plt.grid(False)

    #Convert to dark logo
    dark_logo = 255 - logo1 #Complement every value of the logo np array
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark logo")
    plt.grid(False)
    
    
except FileNotFoundError:
    print("numpy logo file not found")
    
```

pillow has stored the logo in uint8 type of array which is not able to hold pixel data efficiently

### loading the np array created by matplotlib

```python
# 2nd method usingn matplotlib
import matplotlib.image as mpimg
arr2 = mpimg.imread('numpy-logo-png.png')
print(arr2.shape)
print(arr2.dtype)

np.save('2numpy-logo.npy', arr2)
```
