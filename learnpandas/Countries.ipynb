```python
# top 5 country with highest democratic score 
df.sort_values(by='democracy_score', ascending=False).head()
```

```python
# total regions
df['region'].value_counts().count()
```

```python
# number of countries in eastern europe 
df['region'].value_counts()['Eastern Europe']

df[df['region'] == 'Eastern Europe']
```

```python
# leader of the second highest populated country
second_max = df[ df['population'] < df['population'].max() ]['population'].max()
df[ df['population'] == second_max]['political_leader']

# another method
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
# countries with unknown political leaders
df[ df['political_leader'].isna()]
```

```python
# countries having republic in their full name
count = 0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count += 1
    return txt 

df['country_long'] = df['country_long'].apply(counting)
print(count)
```

```python
# country with highest population in african region
df[ (df['continent'] == 'Africa')].sort_values(by='population', ascending=False).iloc[0]
```
