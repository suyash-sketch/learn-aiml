```python
import numpy as np
import pandas as pd
```

### Finding missing data

```python
data = {
    'A' : [1, 2, np.nan, 4, 5],
    'B' : [1, 2, 3, 4, 5],
    'C' : [1, 2, 3, np.nan, np.nan],
    'D' : [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

#### Removing Missing Data

```python
df
```

```python
df.dropna()
```

```python
# Threshold
df.dropna(thresh=3) # there should be atleast 3 non null values (normal regular values) in a row
```

#### Filling the missing data

```python
df
```

```python
df.fillna(0)
```

```python
fill_values = {'A': 0, 'B':100, 'C': 300, 'D' : 400}
df.fillna(value=fill_values)
```

```python
df
```

```python
df.fillna(df.mean())
```
