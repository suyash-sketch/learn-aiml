# Pandas Series
a series is a one-dimensional labeled array capable of holding any datatype. the axis labels are collectively called index

```python
import pandas as pd
import numpy as np
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = { 1:10, 2:20, 3:30 }
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list, index=labels)
```

```python
pd.Series(arr, index=labels)
```

```python
pd.Series(d)
```
