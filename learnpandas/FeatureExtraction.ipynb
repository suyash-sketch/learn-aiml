```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'anime.csv')
```

```python
df.head()
```

### Make a new column for episode count

```python
df.loc[3]['Title']
```

```python
def extract_episode_no(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            return data
            break

        if check == True:
            data += i
        if i == "(":
            check = True
```

```python
df['Episodes'] = df['Title'].apply(extract_episode_no)
```

```python
df.head()
```

```python
df['Episodes'] = df['Episodes'].str.replace(" eps", "")
```

```python
df.head()
```

```python
df['Episodes'] = df['Episodes'].astype(int)
```

```python
df.loc[0]['Episodes']
```

### Make a new column for time stamp

```python
for i in range(50):
    print(df.loc[i]['Title'])
```

```python
def extract_timestamp(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if i == txt.index(')') + 20:
            return data
            break
        
        if txt[i] == ')':
            check = True
            continue
        if check == True:
            data += txt[i]
            
```

```python
df['Title'].apply(extract_timestamp)
```

```python
df['Total_Time'] = df['Title'].apply(extract_timestamp)
```

```python
df.head()
```

```python
def get_total_months(range_str):
    start, end = range_str.split(' - ')
    # print(start)
    # print(end)
    start = pd.to_datetime(start, format='%b %Y')
    end = pd.to_datetime(end, format='%b %Y')
    # print(start)
    # print(end)
    # print((end.year - start.year)*12 + (end.month - start.month))
    return (end.year - start.year)*12 + (end.month - start.month) + 1 # +1 to include the starting month

get_total_months('Feb 1999 - Sep 1999')
```

```python
df['Months'] = df['Total_Time'].apply(get_total_months)
```

```python
df.head()
```

```python
# which anime has the highest score
df[df['Score'] == df['Score'].max()]
```

```python
# anime with highest episode count
df[df['Episodes'] == df['Episodes'].max()]
```

```python
# top 5 anime with highest episode count 
df.sort_values(by='Episodes',ascending= False,axis= 0).head()
```

```python
# longest running anime
df[df['Months'] == df['Months'].max()]
```
