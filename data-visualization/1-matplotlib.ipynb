```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Matplotlib Basics

```python
x = np.linspace(0,5, 11)
x

y = x**2
y
```

```python
plt.plot(x,y)
plt.title('hello')
plt.xlabel('x-axis')
plt.ylabel('y-axix')
```

#### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(x,y)


plt.subplot(2,2,3)
plt.plot(y,x)

plt.subplot(2,2,2)
plt.plot(x**2, x**2)
```

#### **Object oriented way**


Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(8,8), dpi=200) # dpi = dots per inch
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)

axis2 = fig.add_axes([0.16,0.55, 0.4,0.4])
axis2.plot(y,x)
```
