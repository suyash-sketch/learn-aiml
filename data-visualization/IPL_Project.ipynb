```python
# which team won the most matches
match_wins = df['match_winner'].value_counts()
plt.figure(figsize=(12,8))
sns.barplot(x=match_wins.index, y=match_wins.values, palette='viridis')
```

```python
fig = px.bar(x=match_wins.index, y=match_wins.values, color=match_wins.index)
fig.show()
```

```python
# 2. Toss decision head
sns.countplot(x=df['toss_decision'])
```

```python
tosses = df['toss_decision'].value_counts()
tosses
```

```python
px.bar(x=tosses.index, y=tosses.values, labels={ 'x' :'decions', 'y' : 'count'}, color=tosses.index)
```

```python
# 3. Toss Winner vs Match Winner
winner_count = df[df['toss_winner'] == df['match_winner']]['toss_winner'].value_counts()
```

```python
px.bar(x=winner_count.index, y=winner_count.values, color=winner_count.index, labels={'x': 'won toss', 'y' : 'number of times won matches'})
```

```python
# 4. How do teams win( by Runs vs Wicket)
won_by = df['won_by'].value_counts()
px.bar(x=won_by.index, y=won_by.values, labels={'x':'won by','y' : 'wins'}, color=won_by.index)
```

3. 10 Best Bowling Figure

```python
# Extracting the number of wickets from best_bowling_figures

df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)

top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers
px.bar(x=top_bowlers.index, y=top_bowlers.values, color=top_bowlers.values, 
       labels={'x' : 'top_bowlers', 'y' : 'number of wickets'})
```
