```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
fig = px.line(tips, y='total_bill')
fig.show()
```

```python
mean_tip_per_day = tips.groupby('day')['tip'].mean()
mean_tip_per_day
```

```python
fig2 = px.bar(mean_tip_per_day)
fig2.show()
```

```python
tips.iplot(kind='scatter', x='total_bill', y='tip', mode='markers')
```

### **Using plotly-express**

```python
fig3 = px.scatter(tips, x='total_bill', y='tip', color='day')
fig3.show()
```

```python
bill_day_df = tips[['total_bill', 'day']]

fig4 = px.box(tips,x='day', y='total_bill' )
fig4.show()
```
