```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r'../../dataset/breast_cancer.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['diagnosis'].value_counts()
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
```

```python
df.head()
```

```python
# Target and training data

x = df.drop(columns=['diagnosis'])
y = df['diagnosis'] # Target
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y,train_size=0.8, random_state=42)
```

```python
x_train.shape
```

```python
model = Sequential(
    [
        Dense(16, activation='relu', input_dim = 30),
        Dense(8, activation='relu',),
        Dense(1, activation='sigmoid'),
    ]
)

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=150, validation_split=0.2)
```

```python
tr_acc = history.history['accuracy']
tr_loss = history.history['loss']
val_acc = history.history['val_accuracy']
val_loss = history.history['val_loss']

epochs = [i+1  for i in range(len(tr_acc))]
```

```python
len(epochs)
```

```python
len(tr_acc)
len(val_acc)
```

```python
data  = pd.DataFrame( {'train_acc' : tr_acc, 'val_acc' : val_acc}  ,index=epochs)
data2 = pd.DataFrame({'tr_loss': tr_loss, 'val_loss' : val_loss}, index=epochs)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
sns.lineplot(data=data)

plt.xlabel('epoch')
plt.ylabel('loss')

plt.subplot(1,2,2)
sns.lineplot(data=data2)
plt.xlabel('epoch')
plt.ylabel('accuracy')
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
