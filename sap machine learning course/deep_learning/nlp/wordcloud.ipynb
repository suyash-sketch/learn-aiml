```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
a = """Diwali is the five-day festival of lights that celebrates the victory of light over darkness, good over evil, and knowledge over ignorance. Celebrated by Hindus, Jains, and Sikhs around the world, it is a time for families to come together and perform rituals (puja) for deities. A central part of the celebration involves decorating homes, temples, and workplaces with clay oil lamps (diyas), candles, and colorful floor designs known as rangolis. In addition to the stunning light displays, the festival is marked by feasting, sharing homemade sweets (mithai) and gifts with loved ones, and setting off fireworks to signify the triumph of good."""


#Create a wordcloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(a)

#Function to plot the WordCloud
def plot_cloud(wordcloud):
    plt.figure(figsize=(12,5))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()

plot_cloud(wordcloud)
```
