```python
arr1 = pd.Series([1,22,3,4])
arr1
```

```python
arr2 = np.array([1,22,34,46,5])
arr3 = pd.Series(arr2)      #Data must be 1-dimensional
arr3
```

```python
arr3.ndim
```

```python
arr2 = np.array([[1,2,3,4,5]])
arr4 = pd.DataFrame(arr2) # Data must be 2 - dimensional
arr4
```

```python
arr5 = np.array([[1,2,3,4,],
                 [8,9,10,11]])
arr6 = pd.DataFrame(arr5)
arr6
```

```python
aa = {"name" : ["Ajay", "Vijay"], "Grade" : [6.7,8.8]} # asked in interview
aa
```

```python
bb = pd.DataFrame(aa)
bb
```

```python
# you can change the index of the dataFrame but not the dictionary
bb = pd.DataFrame(aa, index=["aa","bb"])
bb
```

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
# Missing value treatment

df.isnull().sum()
```

```python
df1 = df.copy()
df1.head()
```

```python
df1.shape #there are 688 deck missing values out of 891, so deleting them is best option
```

```python
del df1["deck"]
```

```python
df1.columns
```

```python
df1["age"].isnull().sum() #there are 177 deck missing values out of 891, so replace the mssing values with mean
```

```python
df1["age"].mean()
```

```python
df1["age"] = df1["age"].fillna(29.69)
```

```python
df1["age"].isnull().sum()
```

```python
df1.columns
```
