```python
df[df.columns[9:18]].hist(bins=30, figsize=(15,10))
plt.show()
```

```python
plt.figure(figsize=(15,30))
for num, column in enumerate(df.columns):
    if column != 'generated_power_kw':
        plt.subplot(7,3, num + 1)
        plt.scatter(df[column], df['generated_power_kw'])
        plt.xlabel(column)
        plt.ylabel('Generated Power kW')
        plt.title(f'{column} vs Generated Power')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt ='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
salary_df = pd.read_csv(r'dataset/Salary_Data.csv')
salary_df.head()
```

```python
sns.scatterplot(x = salary_df['YearsExperience'], y = salary_df['Salary'])
```

```python
sns.regplot(x=salary_df['YearsExperience'], y=salary_df['Salary'])
```

```python
x = salary_df['YearsExperience'].values.reshape(-1,1)
y = salary_df['Salary'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4, random_state=25)
```

```python
salary_df.shape
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
m = linear.coef_
m
```

```python
c = linear.intercept_
c
```

```python
y_pred = linear.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```
