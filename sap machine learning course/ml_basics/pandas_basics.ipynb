### Finding some Insights

```python
# Survival rates by class

survival_by_class = df1.groupby('Class')['survived'].mean()
print(survival_by_class)
for class_name, rate in survival_by_class.items():
    print(f"{class_name}: {rate:.1%}")
```

```python
# survival rates by gender

survival_by_gender = df1.groupby('sex')['survived'].mean()
print(survival_by_gender.dtype)
print(survival_by_gender)
for gender, rate in survival_by_gender.items():
    print(f"{gender}: {rate: .1%}")
```

```python
# 1. GENDER DISTRIBUTION BY CLASS
print("\n1. GENDER DISTRIBUTION BY CLASS:")
print("-" * 40)

# Count of passengers by class and gender
gender_class = pd.crosstab(df1['Class'], df1['sex'], margins=True)
print("Passenger counts:")
print(gender_class)
print("\n\n")
# Percentage distribution within each class

print("overall eka class madhe kiti males and kiti females, addition rowvise = 100")
gender_class_pct = pd.crosstab(df1['Class'], df1['sex'], normalize='index') * 100
print("\nPercentage within each class:")
print(gender_class_pct.round(1))

# Percentage distribution within each gender
print("\n\n")
print("ek gender madhe konte class che kiti ahet i.e addition column vise = 100")
gender_class_pct2 = pd.crosstab(df1['Class'], df1['sex'], normalize='columns') * 100
print("\nPercentage within each gender:")
print(gender_class_pct2.round(1))
```

```python
age_bins = [0, 12, 18, 25, 30, 40, 50, 80]
age_labels = [
    'Child (0-12)', 
    'Teen (13-18)', 
    'Young Adult (19-25)',
    'Adult (26-30)',
    'Adult (31-40)',
    'Adult (41-50)',
    'Senior (51-80)'
]

df1['AgeGroup'] = pd.cut(df1['age'], bins=age_bins, labels=age_labels, right=True)
```

```python
pd.crosstab(df1['AgeGroup'], df1['Class'])
```

```python
pd.crosstab(df1['AgeGroup'], df1['Class'], normalize="columns") * 100
```

```python
senior_age_bins = [51, 56,61,66,71,76,81]
age_labels = [
    'Senior (51-56)',
    'Senior (57-61)',
    'Senior (62-66)',
    'Senior (67-71)',
    'Senior (72-76)',
    'Senior (77-81)',
]

df1['SeniorAgeGroup'] = pd.cut(df1['age'], bins = senior_age_bins, labels = age_labels, right = True)
```

```python
pd.crosstab(df1['SeniorAgeGroup'], df1['Class'], margins=True)
```

```python
pd.crosstab(df1['SeniorAgeGroup'], df1['Class'], normalize="columns", margins=True) * 100
```
