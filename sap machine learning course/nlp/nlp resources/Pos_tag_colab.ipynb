```python
# 1Ô∏è‚É£ Tokenize the text
tokens = nltk.word_tokenize(sample_text)
```

```python
# 2Ô∏è‚É£ Apply PoS Tagging
pos_tags = nltk.pos_tag(tokens)

print("Tokens with POS tags:")
print(pos_tags)
```

```python
# üîπ VARIATION 2: UNIVERSAL TAGSET (SIMPLER TAGS)
# ============================================================
print("\nüîπ VARIATION 2: Using Universal Tagset\n")

# The universal tagset is much simpler (NOUN, VERB, ADJ, etc.)
pos_tags_universal = nltk.pos_tag(tokens, tagset='universal')

print("Tokens with Universal POS tags:")
print(pos_tags_universal)
```

```python
# üîπ VARIATION 3: USING SPACY FOR POS TAGGING
# ============================================================
print("\nüîπ VARIATION 3: Using spaCy for POS tagging\n")

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")
doc = nlp(sample_text)

for token in doc:
    print(f"{token.text:10} | POS: {token.pos_:10} | Detailed Tag: {token.tag_}")
```

```python
# üîπ VARIATION 4: TRAINING A CUSTOM POS TAGGER (NLTK)
# ============================================================
print("\nüîπ VARIATION 4: Training a very simple custom PoS tagger (for demo)\n")

# ‚úÖ This is a toy example - we use a very tiny training dataset
# Normally you would use a large dataset like Treebank or Brown Corpus

train_data = [
    (["I", "love", "dogs"], ["PRON", "VERB", "NOUN"]),
    (["Dogs", "are", "cute"], ["NOUN", "VERB", "ADJ"]),
    (["She", "hates", "cats"], ["PRON", "VERB", "NOUN"])
]
```

```python
# Prepare training data in the format nltk expects
training_set = []
for words, tags in train_data:
    training_set.append(list(zip(words, tags)))

# Train UnigramTagger (uses only one-word context)
unigram_tagger = nltk.UnigramTagger(training_set)

# Test the custom tagger
test_sentence = ["Dogs", "love", "me"]
print("Custom trained tags:", unigram_tagger.tag(test_sentence))
```

```python
# üîπ VARIATION 5: USING BIGRAM TAGGER WITH BACKOFF
# ============================================================
print("\nüîπ VARIATION 5: Bigram Tagger with Backoff\n")

# Use Unigram Tagger as backoff for Bigram Tagger
bigram_tagger = nltk.BigramTagger(training_set, backoff=unigram_tagger)

print("Bigram + Backoff tags:", bigram_tagger.tag(test_sentence))
```

```python
# üîß INSTALL REQUIRED LIBRARIES
# ===============================================
!pip install nltk spacy matplotlib seaborn --quiet
!python -m spacy download en_core_web_sm

# ===============================================
# üèÅ IMPORT LIBRARIES
# ===============================================
import nltk
import spacy
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
# Download NLTK resources if not already present
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('universal_tagset')
```

```python
# Sample text for PoS Tagging
sample_text = "The quick brown fox jumps over the lazy dog. John loves programming in Python, but Mary prefers Java."
```

```python
# üîπ VARIATION 1: SIMPLE NLTK POS TAGGING
# ============================================================
print("\nüîπ VARIATION 1: Simple PoS Tagging using NLTK\n")

tokens = nltk.word_tokenize(sample_text)
pos_tags = nltk.pos_tag(tokens)
print(pos_tags)
```

```python
# üîπ VARIATION 2: UNIVERSAL TAGSET (SIMPLER TAGS)
# ============================================================
print("\nüîπ VARIATION 2: Using Universal Tagset\n")
pos_tags_universal = nltk.pos_tag(tokens, tagset='universal')
print(pos_tags_universal)
```

```python
# üîπ VARIATION 3: USING SPACY FOR POS TAGGING
# ============================================================
print("\nüîπ VARIATION 3: Using spaCy for POS tagging\n")
nlp = spacy.load("en_core_web_sm")
doc = nlp(sample_text)
for token in doc:
    print(f"{token.text:10} | POS: {token.pos_:10} | Detailed Tag: {token.tag_}")
```

```python
# üîπ VARIATION 4: TRAINING A CUSTOM POS TAGGER
# ============================================================
print("\nüîπ VARIATION 4: Training a very simple custom PoS tagger (for demo)\n")
train_data = [
    (["I", "love", "dogs"], ["PRON", "VERB", "NOUN"]),
    (["Dogs", "are", "cute"], ["NOUN", "VERB", "ADJ"]),
    (["She", "hates", "cats"], ["PRON", "VERB", "NOUN"])
]

training_set = [list(zip(words, tags)) for words, tags in train_data]
unigram_tagger = nltk.UnigramTagger(training_set)
test_sentence = ["Dogs", "love", "me"]
print("Custom trained tags:", unigram_tagger.tag(test_sentence))
```

```python
# üîπ VARIATION 5: USING BIGRAM TAGGER WITH BACKOFF
# ============================================================
print("\nüîπ VARIATION 5: Bigram Tagger with Backoff\n")
bigram_tagger = nltk.BigramTagger(training_set, backoff=unigram_tagger)
print("Bigram + Backoff tags:", bigram_tagger.tag(test_sentence))
```

```python
# üîπ VARIATION 6: VISUALIZATION OF POS TAG FREQUENCIES
# ============================================================
print("\nüîπ VARIATION 6: Visualization of POS Tags\n")

# Get frequency count of tags
tags_only = [tag for word, tag in pos_tags_universal]
tag_counts = Counter(tags_only)

# Plot using Seaborn
plt.figure(figsize=(8,4))
sns.barplot(x=list(tag_counts.keys()), y=list(tag_counts.values()))
plt.title("PoS Tag Frequency")
plt.xlabel("POS Tags")
plt.ylabel("Count")
plt.show()
```
