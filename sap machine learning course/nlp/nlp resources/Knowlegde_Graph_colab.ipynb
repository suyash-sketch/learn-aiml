```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Step 1: Create an empty graph
G = nx.DiGraph() # Directed graph (arrows show direction of relationship)
G
```

```python
# Step 2: Add some nodes (entities)
G.add_node("Apple")
G.add_node("DarkSky")
G.add_node("Tim Cook")
G
```

```python
# Step 3: Add edges (relationships between entities)
# Format: G.add_edge("entity1", "entity2", relation="RELATION_NAME")
G.add_edge("Apple", "DarkSky", relation="ACQUIRED")
G.add_edge("Tim Cook", "Apple", relation="CEO_OF")
```

```python
# Step 4: Visualize the graph
plt.figure(figsize=(6, 4))
```

```python
# Use spring_layout for a nice layout
pos = nx.spring_layout(G, seed=42)
```

```python
# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=800, node_color="lightblue")
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=15)
```

```python
# Draw labels for nodes
nx.draw_networkx_labels(G, pos, font_size=10)
```

```python
# Draw edge labels (relationship names)
edge_labels = {(u, v): d['relation'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red")


plt.axis("off") # Hide axes for clarity
plt.title("Simple Knowledge Graph Example")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def company_graph():
  G = nx.DiGraph()
  G.add_node("Apple")
  G.add_node("DarkSky")
  G.add_node("Tim Cook")
  G.add_edge("Apple", "DarkSky", relation="ACQUIRED")
  G.add_edge("Tim Cook", "Apple", relation="CEO_OF")
  return G, "Company Acquisition Graph"

```

```python
def family_graph():
  G = nx.DiGraph()
  G.add_node("Alice")
  G.add_node("Bob")
  G.add_node("Charlie")
  G.add_edge("Alice", "Bob", relation="MOTHER_OF")
  G.add_edge("Bob", "Charlie", relation="FATHER_OF")
  return G, "Family Relationship Graph"
```

```python
def movie_graph():
  G = nx.Graph() # Undirected graph because friendships are mutual
  G.add_node("Harry")
  G.add_node("Ron")
  G.add_node("Hermione")
  G.add_edge("Harry", "Ron", relation="FRIENDS")
  G.add_edge("Harry", "Hermione", relation="FRIENDS")
  G.add_edge("Ron", "Hermione", relation="FRIENDS")
  return G, "Movie Character Friendship Graph"
```

```python
def visualize_graph(G, title):
  plt.figure(figsize=(6, 4))
  pos = nx.spring_layout(G, seed=42)


  nx.draw_networkx_nodes(G, pos, node_size=800, node_color="lightgreen")
  nx.draw_networkx_edges(G, pos, arrowstyle="->" if isinstance(G, nx.DiGraph) else "-", arrowsize=15)
  nx.draw_networkx_labels(G, pos, font_size=10)


  # Draw edge labels
  edge_labels = {(u, v): d['relation'] for u, v, d in G.edges(data=True)}
  nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red")


  plt.axis("off")
  plt.title(title)
  plt.show()
```

```python
if __name__ == "__main__":
  for graph_func in [company_graph, family_graph, movie_graph]:
    G, title = graph_func()
    print(f"\n--- {title} ---")
    print("Nodes:", list(G.nodes()))
    print("Edges:", list(G.edges(data=True)))
    visualize_graph(G, title)
```
