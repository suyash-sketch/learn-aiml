```python
from nltk.corpus import stopwords
```

```python
import nltk
print(nltk.data.path)
```

```python
nltk.download('stopwords')
```

```python
total_count = stopwords.fileids()
total_count
# len(total_count)
```

```python
stop = stopwords.words('english')
stop
```

```python
text = '''The sigma symbol (Σ, σ) is the 18th letter of the Greek alphabet. 
It has three forms: uppercase "Σ," lowercase "σ," and word-final lowercase "ς". 
In mathematics and statistics, the lowercase sigma (σ) is commonly used to represent standard deviation
'''
```

```python
text1 = word_tokenize(text)
text1
```

```python
stopwords_removed = [ i for i in text1 if i in stop]
stopwords_removed
```

```python
spain_stop = stopwords.words('spanish')
spain_stop
```

```python
spainish_text = """El símbolo sigma (Σ, σ) es la 18ª letra del alfabeto griego. Tiene tres formas: mayúscula "Σ," minúscula "σ," y minúscula final de palabra "ς". En matemáticas y estadísticas, la sigma minúscula (σ) se usa comúnmente para representar la desviación estándar """
```

```python
spainish_text_token = word_tokenize(spainish_text)
spainish_text_token
```

```python
spain_stopword_removed = [ i for i in spainish_text_token if i in spain_stop]
spain_stopword_removed
```
